```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
f = open('logins.json',)
```

```python
data = json.load(f)
```

```python
login_list = (data['login_time'])
```

```python
df = pd.DataFrame(login_list)
```

```python
df.columns = ['login_time']
```

```python
df['login_time']  = pd.to_datetime(df['login_time'] , errors='coerce')  
```

```python
df['count1'] = 1
```

```python
df = df.set_index('login_time')
```

```python
df
```

```python
(df.resample('15min').count())
```

```python
plt.figure(figsize =(20,5))
plt.plot(df.resample('15min').count())
```
